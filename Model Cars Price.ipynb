```python
import pandas as pd
```

```python
car = pd.read_csv("car_data.csv")
```

```python
data = car.copy()
```

```python
target = 'Purchased'
```

```python
encode = ['Gender']
```

```python
data.tail()
```

```python
for col in encode:
    dummy = pd.get_dummies(data[col], prefix=col)
    data = pd.concat([data, dummy], axis=1)
    del data[col]
```

```python
data
```

```python
X = data.drop(['Purchased', 'User ID'], axis=1)
Y = data['Purchased']
```

```python
X
```

```python
Y
```

```python
# Model NBC
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X, Y)
```

```python
# Menyimpan Model NBC
import pickle
pickle.dump(model, open('NBC_car.pkl', 'wb'))
```
